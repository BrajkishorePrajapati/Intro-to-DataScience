## **Matplotlib Architecture**
* Deals with the rendering of plots to screen or files.
* In Jupyter Notebook we use the in line command.<br>

When you run the magic function matplotlib with the inline parameter, is that matplotlib is configured to render into the browser. This configuration is called a backend, and matplotlib has a number of different backends available.<br>
**%matplotlib inline**<br><br>
There are overall 3 different layers in the architecture of matplotlib as follows.<br>
1. Backend layer.
2. Artist layer.
3. Scripting layer.

* **Backend layer** : This is the bottom most layer of a figure which contains implementation of several functions that are required for plotting. There are 3 main classes from the backend layer FigureCanvas ( the layer/Surface on which the figure will be drawn), Renderer (the classn that takes care of the drawing on the surface ) and Event ( to handle the mouse and keyboard events).
+ A backend is an abstraction layer which knows how to interact with the operating environment, whether it's an operating system, or an environment like the browser, and knows how to render matplotlib commands.

* **Artist layer:** The next layer is where we'll spend most of our time though, and that's called the artist layer. The artist layer is an abstraction around drawing and layout primitives. The root of visuals is a set of containers which includes a figure object with one or more subplots, each with a series of one or more axes.

* **Scripting layer:** This is the topmost layer on which majority of our codes will play around. Scripting layer helps simplify and speed up our interaction with the environment in order tot to bulid plots quickly.
* Pyplot is the scripting layer that provides almost simlar functionality as that of Matlab in python<br><br>

### **Line plot :**
* A line plot is generally used to present observations collected at regular intervals.
* A line plot can be created by calling the plot() function and passing the x-axis data for the regular interval, and y-axis for the observations.

```
x = np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y,'-o')  # -o shows a line with dots, --o show the dashed line.
plt.show()  # Display plot
```

**Properties of axis**:
The axis generally have axis to explain what they represent or te units that the describe.
* plt.xlabel('....') - for setting the name of horizontal axis.
* plt.ylabel('....') - for setting the nname of vertical axis.
* plt.title('.....') - for setting the title of the plot.

### **Bar Chart** 
* A bar chart is generally used to present relative quantities for multiple categories.

* The x-axis represents the categories and are spaced evenly. The y-axis represents the quantity for each category and is drawn as a bar from the baseline to the appropriate level on the y-axis.

* A bar chart can be created by calling the bar() function and passing the category names for the x-axis and the quantities for the y-axis.

```
langs = ['C','C++','Python','R']
students_enroll = [20,40,100,15]
plt.bar(langs, students_enroll, width = 0.3)
plt.xlabel('Computer languages',fontsize = 12,fontweight ='bold')
plt.ylabel('Number of students Enroll',fontsize = 12,fontweight ='bold')
plt.title('Students enroll in different languages',fontsize = 14,fontweight ='bold')
plt.show()
```

### **Stacked Bar Graph**

```
langs = ['C','Python','java','php']
students_enroll_it = [20,100,20,19]
students_enroll_cs = [18,80,12,3]
plt.bar(langs, students_enroll_it, width = 0.3, label='IT')

plt.bar(langs, students_enroll_cs, width = 0.3,label = 'CS',color = 'red')
plt.xlabel('Computer languages',fontsize = 12,fontweight ='bold')
plt.ylabel('Number of students Enroll',fontsize = 12,fontweight ='bold')
plt.title('Students enroll in different languages',fontsize = 14,fontweight ='bold')
plt.legend()
plt.show()
```

### **Histogram plot**
* A histogram plot is generally used to summarize the distribution of a data sample.
* The x-axis represents discrete bins or intervals for the observations.The y-axis represents the frequency or count of the number of observations in the dataset that belong to each bin.
* A histogram plot can be created by calling the hist() function and passing in a list or array that represents the data sample.

```
x = np.random.normal(size = 100)
fig = plt.figure(figsize = (6,4))
plt.hist(x,bins = 10) 
plt.show()
```

### **Box and whisker plot**
* Box plot is method of showing aggregate statistics of various samples in a concise manner.
Boxplots are a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum).<br>
 * median (Q2/50th Percentile): the middle value of the dataset.
 * first quartile (Q1/25th Percentile): the middle number between the smallest number (not the minimum) and the median of the dataset.
 * third quartile (Q3/75th Percentile): the middle value between the median and the highest value (not the maximum) of the dataset.
 * interquartile range (IQR): 25th to the 75th percentile.
 * whiskers (shown in blue)
outliers (shown as green circles)
 * maximum: Q3 + 1.5*IQR
 * minimum: Q1 -1.5*IQR

 **Important:** Box and whisker plot is also used to find the outliers.

 image.png

```
from numpy.random import randn
# seed the random number generator
# random numbers drawn from a Gaussian distribution
x = [randn(1000), 5 * randn(1000), 10 * randn(1000)]
# create box and whisker plot
plt.boxplot(x)
# show line plot
plt.show()
```

### **Scatter plot**
* A scatter plot (or scatterplot) is generally used to summarize the relationship between two paired data samples.
* Its main aim two determine how much one variable is affected by another.
* The relationship between two variables is called their correlation.

```
from numpy.random import seed
x = np.random.randn(500)
y = np.random.randn(500)
seed(1)
plt.scatter(x,y)
plt.show()
```

### **Subplots**
* Subplot command allows us to create axis to different partition of this grid.
* It is used to create more than two plots. Suppose you want to create two subplots side by side with parameter 121. It means one row ,two columns and current axis is 1.<br>
Synntax as follows:<br>
plt.subplot(nrows, ncols, currentaxis)

```
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,27])
z= np.array([1,8,27,64,81])

plt.figure()
# 1st line plot
plt.subplot(1,2,1)  # 1,2,1 means 1 row , 2 columns , 1st axis.
plt.plot(x,y,'-o')

# 2nd line plot
plt.subplot(1,2,2)  # 1,2,2 means 1 row , 2 columns , 2nd axis.
plt.plot(x,z,'--o', color ='r')

plt.show()
```

* **Create a 3 by 3 grid of subplots.**

```
# create a 3x3 grid of subplots
fig, ((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9)) = plt.subplots(3, 3, sharex=True, sharey=True)
# plot the linear_data on the 5th subplot axes 
ax5.plot(x,y, '-')
ax1.plot(x,z,'-o')
ax9.plot(y,z,'--')
plt.show()
```

```
# create 2x2 grid of axis subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

# draw n = 10, 100, 1000, and 10000 samples from the normal distribution and plot corresponding histograms
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
plt.show()
```

### * **Filling area between two line plots**

```
#fig = plt.figure(figsize = (8,6))
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,27])
z= np.array([1,8,27,64,81])
plt.plot(x,'-o',y,'-o')

# Filling area between both lines.
plt.gca().fill_between(range(len(x)), x,y, facecolor = 'blue', alpha = 0.25)
plt.show()
```

### *  **Set rotation for x ticklabels**

```
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,27])
z= np.array([1,8,27,64,81])
plt.plot(x,'-o',y,'-o')

# Filling area between both lines.
plt.gca().fill_between(range(len(x)), x,y, facecolor = 'blue', alpha = 0.25)


X = plt.gca().xaxis
for item in X.get_ticklabels():
  item.set_rotation(45)

plt.show()
```

### **Remove frame of the chord**

```
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,27])
z= np.array([1,8,27,64,81])
plt.plot(x,'-o',y,'-o')

# Filling area between both lines.
plt.gca().fill_between(range(len(x)), x,y, facecolor = 'blue', alpha = 0.25)

for spine in plt.gca().spines.values():
  spine.set_visible(False)
plt.show()
```
